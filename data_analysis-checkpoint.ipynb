```python
#  Sessions Trend Over Time (Traffic Growth)

sessions_daily = (
    website_sessions
    .groupby(website_sessions['created_at'].dt.date)
    .size()
    .reset_index(name='sessions')
)

sessions_daily['created_at'] = pd.to_datetime(sessions_daily['created_at'])

plt.figure(figsize=(14,6))
sns.lineplot(data=sessions_daily, x='created_at', y='sessions')
plt.title('Daily Website Sessions Trend', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Number of Sessions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Orders vs Sessions (Conversion Perspective)

orders_daily = (
    orders
    .groupby(orders['created_at'].dt.date)
    .size()
    .reset_index(name='orders')
)

orders_daily['created_at'] = pd.to_datetime(orders_daily['created_at'])

merged_daily = pd.merge(
    sessions_daily,
    orders_daily,
    on='created_at',
    how='left'
).fillna(0)

plt.figure(figsize=(14,6))
plt.plot(merged_daily['created_at'], merged_daily['sessions'], label='Sessions')
plt.plot(merged_daily['created_at'], merged_daily['orders'], label='Orders')
plt.title('Sessions vs Orders Over Time', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Conversion Rate Trend

merged_daily['conversion_rate'] = merged_daily['orders'] / merged_daily['sessions']

plt.figure(figsize=(14,6))
sns.lineplot(data=merged_daily, x='created_at', y='conversion_rate')
plt.title('Daily Conversion Rate Trend', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Conversion Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Revenue Distribution (Clarity on Sales Spread)

plt.figure(figsize=(10,6))
sns.histplot(orders['price_usd'], bins=30, kde=True)
plt.title('Distribution of Order Revenue', fontsize=16)
plt.xlabel('Order Value (USD)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#  Revenue by Product

product_revenue = (
    order_items
    .groupby('product_id')['price_usd']
    .sum()
    .reset_index()
    .merge(products, on='product_id')
)

plt.figure(figsize=(12,6))
sns.barplot(
    data=product_revenue.sort_values('price_usd', ascending=False),
    x='product_name',
    y='price_usd'
)
plt.title('Revenue by Product', fontsize=16)
plt.xlabel('Product')
plt.ylabel('Revenue (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Refund Amount by Product

refunds_by_product = (
    order_item_refunds
    .merge(order_items, on='order_item_id')
    .merge(products, on='product_id')
    .groupby('product_name')['refund_amount_usd']
    .sum()
    .reset_index()
)

plt.figure(figsize=(12,6))
sns.barplot(
    data=refunds_by_product.sort_values('refund_amount_usd', ascending=False),
    x='product_name',
    y='refund_amount_usd'
)
plt.title('Refund Amount by Product', fontsize=16)
plt.xlabel('Product')
plt.ylabel('Refund Amount (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Traffic Source vs Conversion

source_conversion = (
    website_sessions
    .merge(orders, on='website_session_id', how='left')
    .groupby('utm_source')
    .agg(
        sessions=('website_session_id', 'nunique'),
        orders=('order_id', 'nunique')
    )
    .reset_index()
)

source_conversion['conversion_rate'] = source_conversion['orders'] / source_conversion['sessions']

plt.figure(figsize=(12,6))
sns.barplot(data=source_conversion, x='utm_source', y='conversion_rate')
plt.title('Conversion Rate by Traffic Source', fontsize=16)
plt.xlabel('Traffic Source')
plt.ylabel('Conversion Rate')
plt.tight_layout()
plt.show()
```

```python
# Funnel Drop-Off (Sessions → Orders)

total_sessions = website_sessions['website_session_id'].nunique()
total_orders = orders['order_id'].nunique()

funnel_df = pd.DataFrame({
    'stage': ['Sessions', 'Orders'],
    'count': [total_sessions, total_orders]
})

plt.figure(figsize=(8,5))
sns.barplot(data=funnel_df, x='stage', y='count')
plt.title('Website Funnel: Sessions to Orders', fontsize=16)
plt.ylabel('Count')
plt.xlabel('Funnel Stage')
plt.tight_layout()
plt.show()
```

```python
#  Bounce Behavior (Pageviews per Session)

pageviews_per_session = (
    website_pageviews
    .groupby('website_session_id')
    .size()
)

plt.figure(figsize=(10,6))
sns.histplot(pageviews_per_session, bins=30, kde=True)
plt.title('Pageviews per Session Distribution', fontsize=16)
plt.xlabel('Pageviews per Session')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#  Device Type Performance

device_perf = (
    website_sessions
    .merge(orders, on='website_session_id', how='left')
    .groupby('device_type')
    .agg(
        sessions=('website_session_id', 'nunique'),
        orders=('order_id', 'nunique')
    )
    .reset_index()
)

device_perf['conversion_rate'] = device_perf['orders'] / device_perf['sessions']

plt.figure(figsize=(8,5))
sns.barplot(data=device_perf, x='device_type', y='conversion_rate')
plt.title('Conversion Rate by Device Type', fontsize=16)
plt.ylabel('Conversion Rate')
plt.xlabel('Device')
plt.tight_layout()
plt.show()
```

```python
# Revenue vs COGS

profit_daily = (
    orders
    .groupby(orders['created_at'].dt.date)
    .agg(
        revenue=('price_usd', 'sum'),
        cogs=('cogs_usd', 'sum')
    )
    .reset_index()
)

profit_daily['created_at'] = pd.to_datetime(profit_daily['created_at'])

plt.figure(figsize=(14,6))
plt.plot(profit_daily['created_at'], profit_daily['revenue'], label='Revenue')
plt.plot(profit_daily['created_at'], profit_daily['cogs'], label='COGS')
plt.title('Revenue vs COGS Over Time', fontsize=16)
plt.xlabel('Date')
plt.ylabel('USD')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Refund Rate by Product

refund_rate = (
    order_items
    .merge(order_item_refunds, on='order_item_id', how='left')
    .merge(products, on='product_id')
    .groupby('product_name')
    .agg(
        total_sales=('price_usd', 'sum'),
        total_refunds=('refund_amount_usd', 'sum')
    )
    .fillna(0)
    .reset_index()
)

refund_rate['refund_rate'] = refund_rate['total_refunds'] / refund_rate['total_sales']

plt.figure(figsize=(12,6))
sns.barplot(
    data=refund_rate.sort_values('refund_rate', ascending=False),
    x='product_name',
    y='refund_rate'
)
plt.title('Refund Rate by Product', fontsize=16)
plt.xlabel('Product')
plt.ylabel('Refund Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  User Repeat Behavior

repeat_dist = website_sessions['is_repeat_session'].value_counts().reset_index()
repeat_dist.columns = ['is_repeat_session', 'count']

plt.figure(figsize=(6,6))
plt.pie(
    repeat_dist['count'],
    labels=['New Sessions', 'Repeat Sessions'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('New vs Repeat Sessions')
plt.tight_layout()
plt.show()
```

```python
# Top Landing Pages

landing_pages = (
    website_pageviews
    .sort_values('created_at')
    .groupby('website_session_id')
    .first()
    .reset_index()
)

top_landing_pages = (
    landing_pages['pageview_url']
    .value_counts()
    .head(10)
    .reset_index()
)

top_landing_pages.columns = ['page', 'sessions']

plt.figure(figsize=(12,6))
sns.barplot(data=top_landing_pages, x='page', y='sessions')
plt.title('Top Landing Pages', fontsize=16)
plt.xlabel('Landing Page')
plt.ylabel('Sessions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cohort Analysis (User Retention Over Time)

sessions_with_date = website_sessions.copy()
sessions_with_date['session_date'] = sessions_with_date['created_at'].dt.to_period('M')

user_cohort = (
    sessions_with_date
    .groupby('user_id')['session_date']
    .min()
    .reset_index()
    .rename(columns={'session_date': 'cohort_month'})
)

sessions_with_date = sessions_with_date.merge(user_cohort, on='user_id')
```

```python
cohort_data = (
    sessions_with_date
    .groupby(['cohort_month', 'session_date'])['user_id']
    .nunique()
    .reset_index()
)

cohort_data['cohort_index'] = (
    cohort_data['session_date'].astype(int)
    - cohort_data['cohort_month'].astype(int)
)
```

```python
cohort_pivot = cohort_data.pivot_table(
    index='cohort_month',
    columns='cohort_index',
    values='user_id'
)

plt.figure(figsize=(12,6))
sns.heatmap(cohort_pivot, annot=True, fmt='.0f', cmap='Blues')
plt.title('User Retention Cohort Analysis')
plt.xlabel('Months Since First Visit')
plt.ylabel('Cohort Month')
plt.tight_layout()
plt.show()
```

```python
#  RFM Analysis (Customer Value Segmentation)

snapshot_date = orders['created_at'].max() + pd.Timedelta(days=1)

rfm = (
    orders
    .groupby('user_id')
    .agg(
        recency=('created_at', lambda x: (snapshot_date - x.max()).days),
        frequency=('order_id', 'count'),
        monetary=('price_usd', 'sum')
    )
    .reset_index()
)

rfm.describe()
```

```python
# Time-to-Convert Analysis (Behavior Insight)

session_order_time = (
    website_sessions
    .merge(orders, on='website_session_id', how='inner')
)

session_order_time['time_to_order_minutes'] = (
    (session_order_time['created_at_y'] - session_order_time['created_at_x'])
    .dt.total_seconds() / 60
)

plt.figure(figsize=(10,6))
sns.histplot(session_order_time['time_to_order_minutes'], bins=50, kde=True)
plt.title('Time to Conversion (Minutes)')
plt.xlabel('Minutes')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Channel ROI Analysis

channel_revenue = (
    website_sessions
    .merge(orders, on='website_session_id', how='left')
    .groupby('utm_source')
    .agg(
        sessions=('website_session_id', 'nunique'),
        revenue=('price_usd', 'sum')
    )
    .reset_index()
)

channel_revenue['revenue_per_session'] = channel_revenue['revenue'] / channel_revenue['sessions']

plt.figure(figsize=(12,6))
sns.barplot(data=channel_revenue, x='utm_source', y='revenue_per_session')
plt.title('Revenue per Session by Channel')
plt.xlabel('Channel')
plt.ylabel('Revenue per Session')
plt.tight_layout()
plt.show()
```

```python
# Product Attachment / Cross-Sell Analysis

order_product_counts = (
    order_items
    .groupby('order_id')['product_id']
    .nunique()
)

plt.figure(figsize=(10,6))
sns.histplot(order_product_counts, bins=10)
plt.title('Number of Products per Order')
plt.xlabel('Products per Order')
plt.ylabel('Orders')
plt.tight_layout()
plt.show()
```

```python
# Seasonality Analysis (Monthly Trends)

orders['month'] = orders['created_at'].dt.month

monthly_revenue = (
    orders
    .groupby('month')['price_usd']
    .sum()
    .reset_index()
)

plt.figure(figsize=(10,6))
sns.lineplot(data=monthly_revenue, x='month', y='price_usd', marker='o')
plt.title('Monthly Revenue Seasonality')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis (Numerical Relationships)

corr_df = orders[['price_usd', 'cogs_usd', 'items_purchased']]

plt.figure(figsize=(6,5))
sns.heatmap(corr_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Order Metrics')
plt.tight_layout()
plt.show()
```
